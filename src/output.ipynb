```python
import data
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data_dir = "../data/"
files = data.get_files("../data/")
```

```python
for f in files:
    G = data.get_graph(data_dir+f)
    n = len(G.nodes())
    e = len(G.edges())
    print("{} has {} nodes, {} edges.".format(f,n,e))
```

# PageRank

```python
import pagerank
```

```python
PR = []
for f in files:
    print(f)
    G = data.get_graph(data_dir+f)
    pr = pagerank.pagerank(G)
    print(pr)
    pagerank.output_file(pr,data_dir+f)
    PR.append(pr)
```

# HITS

```python
import HITS
HUB = []
AUT = []
for f in files:
    print(f)
    G = data.get_graph(data_dir+f)
    aut,hub = HITS.HITS(G)
    print("Authority:")
    print(aut)
    print("Hub:")
    print(hub)
    HITS.output_file(aut,hub,data_dir+f)
    HUB.append(hub)
    AUT.append(aut)
```

# SimRank

```python
import Simrank
SR = []
for f in files[:-2]:
    print(f)
    G = data.get_graph(data_dir+f)
    sr = Simrank.Simrank(G)
    print(sr)
    Simrank.output_file(sr,data_dir+f)
    SR.append(sr)
```

# Final Output

```python
for i,f in enumerate(files):
    print(f)
    print("PageRank:")
    print(PR[i])
    print("Authority:")
    print(AUT[i])
    print("Hub:")
    print(HUB[i])
    if i<len(SR):
        print("SimRank:")
        print(SR[i])
    print()
```

# Find a Way
Find a way (e.g., add/delete some links) to increase hub, authority, and PageRank of Node 1 in first 3 graphs respectively

```python
for f in files[:3]:
    print(f)
    G = data.get_graph(data_dir+f)
    
    # Original
    pr = pagerank.pagerank(G)[0]
    aut,hub = HITS.HITS(G)
    aut = aut[0]
    hub = hub[0]
    
    # Way: Add 2 edges, which 1 in and 1 out
    ## Acknowledge: These 3 graphs does not have edge 3->1 and 1->4
    n = len(list(G.nodes))
    n_G = G.copy()
    n_G.add_edge(3,1)
    n_G.add_edge(1,4)
    
    n_pr = pagerank.pagerank(n_G)[0]
    n_aut,n_hub = HITS.HITS(n_G)
    n_aut = n_aut[0]
    n_hub = n_hub[0]
    
    print()
    print("Hub: {:0.3f} -> {:0.3f}".format(hub,n_hub))
    print("Authority: {:0.3f} -> {:0.3f}".format(aut,n_aut))
    print("PageRank: {:0.3f} -> {:0.3f}".format(pr,n_pr))
    print()
    subax1 = plt.subplot(121)
    nx.draw(G,with_labels = True,font_weight='bold')
    subax2 = plt.subplot(122)
    nx.draw(n_G,with_labels = True,font_weight='bold')
    plt.show()
```

# Result Analysis and Discussion

```python
for f in files[:3]:
    print(f)
    G = data.get_graph(data_dir+f)
    
    # Original
    aut,hub = HITS.HITS(G)
    aut = aut
    hub = hub
    sr = Simrank.Simrank(G)
    
    # Different damping factor
    d = [0.1,0.125,0.15]
    pr = [pagerank.pagerank(G,d = d[0]),
          pagerank.pagerank(G,d = d[1]),
          pagerank.pagerank(G,d = d[2])]
    
    # Differenet delay factor
    C = [0.7,0.8,0.9]
    sr = [Simrank.Simrank(G,C = C[0]),
          Simrank.Simrank(G,C = C[1]),
          Simrank.Simrank(G,C = C[2])]
    
    print()
    print("Hub: {}".format(hub))
    print("Authority: {}".format(aut))
    print("PageRank(d={}): {}".format(d[0],pr[0]))
    print("PageRank(d={}): {}".format(d[1],pr[1]))
    print("PageRank(d={}): {}".format(d[2],pr[2]))
    print("SimRank(C={}):\n {}".format(C[0],sr[0]))
    print("SimRank(C={}):\n {}".format(C[1],sr[1]))
    print("SimRank(C={}):\n {}".format(C[2],sr[2]))
    print()
```
