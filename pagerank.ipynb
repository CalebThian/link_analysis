```python
d=0.1
path = './data/graph_5.txt'
```

```python
def PageRank(adj_mat,n,e):
    last_pr = np.ones(n)/n
    pr = last_pr.copy()
    outdegree = np.sum(adj_mat,axis = 1)
    eps = 0.0001
    while True:
        '''
        for i in range(n):
            par = np.where(adj_mat.transpose()[i]==1)[0]
            pr_in = 0
            if len(par) != 0:
                for p in par:
                    pr_in += last_pr[p]/outdegree[p]
            print(pr_in)
            pr[i] = d/n+(1-d)*pr_in
        pr = pr/np.sum(pr)
        '''
        
        for i in range(n):
            dampling = adj_mat.transpose()[i]*last_pr.copy()
            for j in range(n):
                if dampling[j] != 0:
                    dampling[j] = dampling[j]/outdegree[j]
            pr[i] = d/n + (1-d)*np.sum(dampling)
        pr = pr/np.sum(pr)
        if np.linalg.norm(pr-last_pr) < eps:
            break
        else:
            last_pr = pr.copy()
    print(pr)
    return pr
        
```
